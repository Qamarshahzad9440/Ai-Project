```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
# plt.show()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D Diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
# plt.xlabel('x axis')
# plt.ylabel('y axis')
# plt.title('2D Diagram')
plt.subplot(2,2,2)
plt.plot(x,y,'b*--')
# plt.xlabel('x axis')
# plt.ylabel('y axis')
# plt.title('2D Diagram')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
# plt.xlabel('x axis')
# plt.ylabel('y axis')
# plt.title('2D Diagram')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('Matplotlib Demo')
plt.xlabel('X axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.title('Sine wave form')

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')
          
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')
plt.show()
```

# Pie Chart

```python
labels='Python','C++','Rubby','Java'
sizes=[215,130,245,210]
colors=['Gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
