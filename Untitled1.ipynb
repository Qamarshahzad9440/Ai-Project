```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.dtypes
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

# jointplot 

```python
sns.jointplot(x='sepal_length',y='petal_length', data=iris,kind='hex')
```

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='reg')
```

# pairplot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='petal_length')
```

# distplot

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.distplot(iris["sepal_length"],kde=False,bins=10)
```

# CetaGorical Plots
Contplots
Boxplot
Barplot

```python
sns.countplot('sepal_length',data=iris)
```

```python
sns.countplot('sepal_width',data=iris)
```

```python
sns.countplot(y='sepal_width',data=iris)
```

```python
sns.barplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.barplot(y='sepal_width',x='petal_width',data=iris)
```

```python
sns.boxplot('petal_length','sepal_width',data=iris)
```

```python
sns.boxplot(x='sepal_length',y='petal_length',data=iris, palette="rainbow")
```

```python
sns.boxplot(data=iris,orient='v')
```

```python
sns.violinplot(x='petal_width',y='sepal_length',data=iris)
```
